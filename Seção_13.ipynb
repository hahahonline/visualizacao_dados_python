```python
#importação das bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as srn
```

```python
#carregamento da base de dados
mun_gaucho = pd.read_csv('dados.csv', sep=';')
mun_gaucho.shape #isso aqui vai retornar as dimensões do dataframe (linhas e colunas)
```

```python
#cria um histograma com os valores presentes no pib
#kde é estimativa da densidade kernel, é basicamente essa "linha" que fica oscilando no gráfico
#bins é o numero de barras do histograma
srn.histplot(mun_gaucho.PIB, kde=True, bins=10).set(title='PIB')
```

```python
srn.histplot(mun_gaucho.VALOREMPENHO, kde=True, bins=10).set(title="Empenho")
```

```python
agrupado = mun_gaucho.sort_values('VALOREMPENHO').head(10)
agrupado = agrupado.iloc[:,[1,3]]
agrupado
agrupado.plot.bar(x='MUNICIPIO',y='VALOREMPENHO', color = 'gray')
```

```python
agrupado = mun_gaucho.sort_values('VALOREMPENHO').head(10)
agrupado = agrupado.iloc[:, [1, 3]]

# cria o gráfico
plt.figure(figsize=(10, 6))
srn.barplot(x='VALOREMPENHO', y='MUNICIPIO', data=agrupado, color='gray')

# configs adicionais ->
plt.xlabel('Valor Empenho')
plt.ylabel('Município')
plt.title('Empenho de 10 Municípios do RS')
plt.grid(True)
# <-

plt.show()
```

```python
agrupado = mun_gaucho.sort_values('PIB').head(10)
agrupado = agrupado.iloc[:,1:3]


# cria o gráfico
plt.figure(figsize=(10, 6))
srn.barplot(x='PIB', y='MUNICIPIO', data=agrupado, color='gray')

# configs adicionais ->
plt.xlabel('PIB')
plt.ylabel('Município')
plt.title('PIB de 10 Municípios do RS')
plt.grid(True)
# <-

plt.show() 
```
