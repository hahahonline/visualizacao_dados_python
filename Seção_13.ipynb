```python
# Importação das bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as srn
```

```python
# Carregamento da base de dados
mun_gaucho = pd.read_csv('dados.csv', sep=';')
mun_gaucho.shape #isso aqui vai retornar as dimensões do dataframe (linhas e colunas)
```

```python
#cria um histograma com os valores presentes no pib
#kde é estimativa da densidade kernel, é basicamente essa "linha" que fica oscilando no gráfico
#bins é o numero de barras do histograma
srn.histplot(mun_gaucho.PIB, kde=True, bins=10).set(title='PIB')
```

```python
srn.histplot(mun_gaucho.VALOREMPENHO, kde=True, bins=10).set(title="Empenho")
```
